# Tratando valores em Branco (NaN)
## Econtrando e preenchendo valores em branco

```python
import pandas as pd
import numpy as np

from pandas import Series, DataFrame
```

```python
#Encontrando valores em branco
em_branco = np.nan # utilizando numpy para encontrar valores em branco

serie = Series(['linha 1', 'linha 2', em_branco, 'linha4', 'linha 5', 'linha 6', em_branco,'linha 8'])
serie
```

```python
# método .isnull retorna valores booleanos ao comparar os itens dentro do objeto como nan, do numpy
serie.isnull()
```

## preenchendo valores em branco

```python
#Criando um dataframe com valores randomicos
np.random.seed(25)
df = DataFrame(np.random.rand(36).reshape((6,6)))
df.loc[3:5, 0] = em_branco # alterando valores das linhas 3 a 5, da coluna 0
df.loc[1:4, 5] = em_branco # alterando valores das linhas 1 a 3, da coluna 5
df
```

```python
# método .fillna para preencher os valores vazios, esse método retorna sempre um novo dataframe
df_preenchido = df.fillna(0) #passei como parâmetro o valor com que quero preencher
df_preenchido
```

```python
# também dá pra utilizar o .fillna para atribuir valores diferentes para colunas diferentes 
dicio = {0: 0.1, 5: 1.25} # este dicionário recebe o nome da coluna e o valor de atribuição
df_preenchido = df.fillna(dicio)
df_preenchido
```

```python
# preenchido com a média das colunas
media = df.mean(0,1) # média das colunas
df_preenchido = df.fillna(media)
df_preenchido
```

## Removendo dados duplicados

```python
df.drop_duplicates(inplace=True) # utilizar o inplace para salvar a alteração no df original
df
```

# Concatenando Dados


```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
df3 = pd.DataFrame(np.arange(36).reshape(6,6))
df3
```

```python
df4 = pd.DataFrame(np.arange(15).reshape(5,3))
df4
```

```python
# pd.concat([obj-da-esquerda, objt-da-direita])

pd.concat([df3, df4]) #pega as linhas do segundo objeto e coloca abaixo das linhas do primeiro objeto, ou seja, considera os indices das colunas
```

```python
# pd.concat([obj-da-esquerda, objt-da-direita], axis=1) força a juntar os dados considerando os indices das linhas
pd.concat([df3, df4],axis=1)
```

## Transformando dados
* remove dados - eliminando vairáveis ou observações
* adiciona dados - acrescentando variáveis ou observações
* Ordena dados 

```python
import numpy as np
import pandas as pd
```

```python
df5 = pd.DataFrame(np.arange(36).reshape(6,6))
df5
```

## Removendo dados

```python
#retirando linhas através dos indices das linhas a ser removidas
df5.drop([0, 2]) #retorna um novo df
```

```python
#retirando colunas
df5.drop([1, 2], axis=1)
```

## Acrescentando dados

```python
serie = pd.Series(np.arange(6))
serie.name = 'nova_variavel'
serie
```

```python
# adicionar a serie como uma variável no df
novo_df5 = pd.DataFrame.join(df5, serie)
novo_df5
```

## Ordenando dados

```python
# ordena pela coluna nova_variavel de forma decrescente
df_ordenado = novo_df5.sort_values(by=['nova_variavel'], ascending=False)
df_ordenado 
```

# Agrupamentos e agregações


```python
import pandas as pd
```

```python
#caminho = "C:/Users/Dheni/Documents/Cursos TI/python para ciencia de dados/dados/mtcars.csv"
# dados relativos ao consumo de combustível
caminho = "/content/mtcars.csv"

carros = pd.read_csv(caminho)
carros.head()
```

```python
# Renomeando colunas
#todo df possui um atributo chamado columns, esse atributo contém os nomes das colunas
carros.columns = ['nome',	'mpg', 'cyl', 'disp', 'hpt', 'drat',	'wt',	'qsec',	'vs',	'am', 'gear', 'carb'] 
carros.head()
```

## Agrupamentos dados por uma coluna

```python
# .groupby('nome-da-serie')
# agrupando os carros por cilindros (cyl)
coluna_agrupamento = carros['cyl']
grupos_carros = carros.groupby(coluna_agrupamento)

#calcula a média das outras colunas
grupos_carros.mean()
```
